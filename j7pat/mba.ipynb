```python
# Perform the Apriori algorithm on the preprocessed dataset
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

# Filter the results based on minimum support and length of itemsets
filtered_itemsets = frequent_itemsets[(frequent_itemsets['length'] == 2) & (frequent_itemsets['support'] >= 0.01)]
print("Apriori Algorithm:")
print(filtered_itemsets)
```

```python
# Add implementation of FP-Growth algorithm
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets_fp = fpgrowth(df, min_support=0.05, use_colnames=True)
frequent_itemsets_fp['length'] = frequent_itemsets_fp['itemsets'].apply(lambda x: len(x))
print("FP-Growth Algorithm:")
print(frequent_itemsets_fp)
```

```python
# Add implementation of Eclat-based algorithm
from mlxtend.frequent_patterns import association_rules
association_rules_eclat = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
print("Eclat-based Algorithm:")
print(association_rules_eclat)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('mba.csv')

# List of items to label
items_to_label = ['shrimp','almonds','avocado','vegetables mix','green grapes','whole weat flour','yams','cottage cheese','energy drink','tomato juice','low fat yogurt','green tea','honey','salad','mineral water','salmon','antioxydant juice','frozen smoothie','spinach','olive oil']


# Create a list to store the labeled data
labeled_data = []

# Iterate over each transaction
for index, row in df.iterrows():
    labeled_items = {}
    
    # Mark the presence of items in the transaction
    for item in items_to_label:
        if row[item] == 1:  # Assuming 1 indicates the presence of an item in the column
            labeled_items[item] = 1
        else:
            labeled_items[item] = 0
    
    # Append the labeled items to the list
    labeled_data.append(labeled_items)

# Convert the list to a DataFrame
labeled_df = pd.DataFrame(labeled_data)

# Save the labeled data to a new CSV file
labeled_df.to_csv('labeled_data.csv', index=False)
```
